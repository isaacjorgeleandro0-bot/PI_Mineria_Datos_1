#05-CONCLUSIONES

El análisis realizado sobre el conjunto de datos de usuarios de una plataforma de streaming permitió transformar una base con problemas de calidad en un conjunto de datos confiable y apto para el análisis estadístico y la aplicación de técnicas de aprendizaje automático.

Durante la etapa de inspección inicial se identificaron inconsistencias importantes, entre ellas errores de escritura en variables categóricas, registros duplicados, valores faltantes, datos atípicos y formatos incorrectos. Estas anomalías evidenciaron la necesidad de un proceso de limpieza previo antes de realizar cualquier análisis.

En la fase de calidad y limpieza de datos se aplicaron procedimientos de estandarización, eliminación de duplicados, tratamiento de valores nulos e imputaciones justificadas según el comportamiento esperado de cada variable. También se corrigieron valores extremos mediante técnicas de capping y reemplazo por medidas estadísticas cuando correspondía. Como resultado, se obtuvo un conjunto de datos consistente, sin registros duplicados relevantes ni valores faltantes, preservando la mayor cantidad posible de información.

El análisis exploratorio de datos permitió comprender el comportamiento de los usuarios de la plataforma. Se observó que la mayor parte de los clientes mantiene un nivel de consumo mensual elevado, lo que indica un buen nivel general de engagement. Asimismo, la distribución de los planes de suscripción mostró una estructura piramidal, donde el plan básico concentra la mayor cantidad de usuarios.

En el análisis bivariado se comprobó que los usuarios con planes Premium presentan un tiempo de visualización considerablemente mayor que aquellos con planes inferiores, lo que respalda la propuesta de valor de las suscripciones de mayor costo. En contraste, no se encontraron evidencias de una relación significativa entre la edad de los usuarios y la cantidad de tickets generados al soporte técnico, indicando que la utilización de la plataforma no parece depender de la edad desde el punto de vista de la asistencia técnica.

El análisis multivariado permitió identificar un perfil potencialmente asociado con un mayor riesgo de abandono: usuarios de mayor edad suscriptos al plan Básico y con bajos niveles de consumo de contenido. Este hallazgo representa una oportunidad para que la empresa implemente estrategias de fidelización específicas dirigidas a este segmento.

Finalmente, el Análisis de Componentes Principales (PCA) permitió reducir la dimensionalidad de las variables numéricas, sintetizando gran parte de la información original en un conjunto reducido de componentes principales. La primera componente se relacionó principalmente con el nivel de actividad y la interacción del usuario con la plataforma, mientras que las restantes capturaron variaciones complementarias del comportamiento. Esta transformación facilita futuras tareas de modelado, segmentación y visualización sin perder una proporción significativa de la información contenida en los datos.

En conjunto, el proyecto demuestra la importancia de un adecuado proceso de preparación de datos previo al análisis. La calidad alcanzada en el dataset permitió obtener resultados confiables y generar conocimiento útil para la toma de decisiones, especialmente en aspectos relacionados con la retención de clientes, la optimización de planes de suscripción y la comprensión de los patrones de consumo de los usuarios.
